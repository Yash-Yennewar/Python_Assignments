```python
# 1. Seconds in an Hour
# Use Python to calculate how many seconds are in one hour. 
# Multiply the number of seconds in a minute by the number of minutes in an hour.

seconds_per_minute = 60
minutes_per_hour = 60
seconds_per_hour = seconds_per_minute * minutes_per_hour

# 2. Store the Result in a Variable
# Assign the result from Question 1 to a variable called seconds_per_hour and print it.

print("Seconds in one hour:", seconds_per_hour)
```

```python
# 3. Calculate Seconds in a Day
# Estimate how many seconds are in a day. Use the variable seconds_per_hour and any other variables you may need.

hours_per_day = 24
seconds_per_day = seconds_per_hour * hours_per_day

print("Seconds in one day:", seconds_per_day)
```

```python
# 4. Save Seconds Per Day in a Variable
# Recalculate the number of seconds in a day and assign the result to a variable called seconds_per_day. Print the value.

seconds_per_hour = 3600
hours_per_day = 24
seconds_per_day = seconds_per_hour * hours_per_day

print("Seconds in one day:", seconds_per_day)
```

```python
# 5. Floating-Point Division
# Divide seconds_per_day by seconds_per_hour using floating-point division (/). What result do you get?

hours_in_day = seconds_per_day / seconds_per_hour
print("Hours in a day (floating-point division):", hours_in_day)

# / always gives a float result in Python.
# 86400 รท 3600 = 24.0
```

```python
# 6. Integer Division
# Now divide seconds_per_day by seconds_per_hour using integer division (//).
# Does this value match the previous result aside from the decimal point?

hours_in_day_int = seconds_per_day // seconds_per_hour
print("Hours in a day (integer division):", hours_in_day_int)

# // gives the integer quotient (no decimal).
# The value matches the previous result aside from the decimal point: 24.0 (float) vs 24 (int).
```

```python
# 7. Prime Number Generator
# Write a generator function genPrimes() that yields an infinite sequence of prime numbers starting from 2. 
# Use a loop to print the first 5 prime numbers generated by genPrimes().

def genPrimes():
    primes = []
    num = 2
    while True:
        is_prime = True
        for p in primes:
            if num % p == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
            yield num
        num += 1

prime_gen = genPrimes()
for _ in range(5):
    print(next(prime_gen),end=" ")
```
